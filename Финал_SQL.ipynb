## Исследование

### Посчитайте, сколько книг вышло после 1 января 2000 года

```python
query='''
select count(book_id) from books
where publication_date > '1/1/2000'
limit 20
'''
pd.io.sql.read_sql(query, con = engine)
```

После 1 января 2000 года было выпущено 819 книг

###  Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query='''
select title, book_id, avg(ratings.rating), count(distinct reviews.review_id) from books
left join ratings using (book_id)
left join reviews using (book_id)
group by books.book_id
order by count(distinct reviews.review_id)  desc
'''
pd.io.sql.read_sql(query, con = engine)
```

Наибольшее количество обзоров имеет книга Twilight (7), большинство оценок колеблются от 3.5 до 4.5 

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query='''

select count(book_id) c, publisher  from books
join publishers using (publisher_id)
where num_pages>50
group by publisher
order by c desc
limit 1
'''

pd.io.sql.read_sql(query, con = engine)
```

Наибольшее число книг выпустило издательство Penguin Books - 42.

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query='''
select author,
       round(avg(rating), 2) as avg_rating
from authors as a
join books as b on b.author_id=a.author_id
join ratings as r on r.book_id=b.book_id
where b.book_id in (select  book_id
                            from ratings
                            group by book_id
                            having count(rating) >= 50)
group by author
order by avg(rating) desc
limit 1

'''
pd.io.sql.read_sql(query, con = engine)
```

J.K. Rowling  имеет самую высокую среднюю оценку - 4.29.

### Посчитайте среднее количество обзоров от пользователей, которые поставили более чем по 50 оценок

```python
query = '''
select round(avg(count)) 
from (select count(text)
from reviews
join (select username, count(rating_id) as cnt
from ratings
group by username) as cnt_rating
on cnt_rating.username = reviews.username
where cnt > 50
group by reviews.username) as count;                                  
'''


pd.io.sql.read_sql(query, con = engine)
```

Среднее количество обзоров - 24
